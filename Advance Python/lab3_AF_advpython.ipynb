```python
# ques1
# calculate total revenue generated by 2 product stores
import numpy as np
pro1=np.array([500,600,700,550])
pro2=np.array([450,700,800,600])
print(pro1+pro2)
```

```python
#ques2
# calculate profit
revenue=np.array([10000,12000,11000,10500])
expense=np.array([4000,5000,4500,4800])
print(revenue-expense)
```

```python
# ques3
# calculate items which are unavailable
inventory=np.array([10,0,5,0,20,0])
print(*np.where(inventory==0))
```

```python
# ques4
# calculate total cost of items
qty=np.array([2,3,4,1])
price_per_item=np.array([10.0,5.0,8.0,12.0],dtype='int')       
print(qty*price_per_item)
```
