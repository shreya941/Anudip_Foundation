```python
#Lab4: Suppose you're a sales manager for an e-commerce company, and you want to create a figure with subplots to compare the sales performance of different product categories over time. You have sales data for four productcategories: Electronics, Clothing, Home & Garden, and Sports & Outdoors. Share your conclusion/analysis. 

months = np.arange(1, 13)
electronics_sales = np.array([25000, 28000, 31000, 27000, 30000, 32000, 35000, 36000, 38000, 39000, 41000, 42000])
clothing_sales = np.array([15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000, 23000, 24000, 25000, 26000])
home_garden_sales = np.array([18000, 19000, 20000, 21000, 22000, 23000, 24000, 25000, 26000, 27000, 28000, 29000])
sports_outdoors_sales = np.array([12000, 13000, 14000, 15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000, 23000])


fig, axs = plt.subplots(2, 2, figsize=(7,5))
fig.suptitle('Sales Performance Comparison', fontsize=16)


axs[0, 0].plot(months, electronics_sales, linestyle='-', color='b', marker='o')
axs[0, 0].set_title('Electronics')
axs[0, 0].set_xlabel('Months')
axs[0, 0].set_ylabel('Sales')


axs[0, 1].plot(months, clothing_sales, linestyle='-', color='g', marker='o')
axs[0, 1].set_title('Clothing')
axs[0, 1].set_xlabel('Months')
axs[0, 1].set_ylabel('Sales')

axs[1, 0].plot(months, home_garden_sales, linestyle='-', color='r', marker='o')
axs[1, 0].set_title('Home & Garden')
axs[1, 0].set_xlabel('Months')
axs[1, 0].set_ylabel('Sales')

axs[1, 1].plot(months, sports_outdoors_sales, linestyle='-', color='purple', marker='o')
axs[1, 1].set_title('Sports & Outdoors')
axs[1, 1].set_xlabel('Months')
axs[1, 1].set_ylabel('Sales')

plt.tight_layout(rect=[0, 0, 1, 0.96])


plt.show()
```

```python
#Suppose you work for a retail company, and you have dummy data containing sales data for the past year. The data includes information such as SalesDate,product names,regions, sales quantities, prices, and dates. You have to generate a bar chart ,pie plot on region and prices and line chart on SalesDate and prices columns. Further, you need to get some inference out of the chart

import numpy as np
import matplotlib.pyplot as plt

# Dummy Data
months = np.arange(1, 13)  # Representing months 1 to 12
products = ['Electronics', 'Clothing', 'Home & Garden', 'Sports & Outdoors'] * 3
regions = ['North', 'South', 'East', 'West'] * 3
sales_quantities = np.random.randint(50, 200, size=12)
prices = np.random.uniform(100, 500, size=12)

# Aggregating data
unique_products = np.unique(products)
product_sales = [np.sum(sales_quantities[np.array(products) == product]) for product in unique_products]

unique_regions = np.unique(regions)
region_sales = [np.sum(prices[np.array(regions) == region]) for region in unique_regions]

# Since we are not using dates, let's use months and average prices per month
price_per_month = [np.mean(prices[i:i+1]) for i in range(len(months))]  # Mean price for each month

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Bar Chart: Sales Quantity by Product
axs[0, 0].bar(unique_products, product_sales, color='skyblue')
axs[0, 0].set_title('Sales Quantity by Product')
axs[0, 0].set_xlabel('Product')
axs[0, 0].set_ylabel('Sales Quantity')

# Pie Chart: Sales by Region
axs[0, 1].pie(region_sales, labels=unique_regions, autopct='%1.1f%%', startangle=140)
axs[0, 1].set_title('Sales Distribution by Region')

# Line Chart: Prices over Months
axs[1, 0].plot(months, prices, marker='o', linestyle='-', color='purple')
axs[1, 0].set_title('Price Over Months')
axs[1, 0].set_xlabel('Month')
axs[1, 0].set_ylabel('Price')

# Pie Chart: Price Distribution
unique_prices, price_counts = np.unique(prices, return_counts=True)
axs[1, 1].pie(price_counts, labels=unique_prices.round(2), autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Price Distribution')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show plot
plt.show()
```

Inferences:
1. Bar Chart: Highlights which products are performing better based on sales quantities.
2. Pie Chart (Region): Shows which regions are contributing more to total sales.
3. Line Chart: Provides insights into how prices fluctuate or trend across different months.
4. Pie Chart (Price Distribution): Visualizes the frequency of different price points, indicating common price ranges.
