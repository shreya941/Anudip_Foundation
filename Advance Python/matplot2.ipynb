```python
import matplotlib.pyplot as plt

students=['John','Shreya','Yashika','Sophia','Marry']
maths=[22,23,25,21,20]
science=[22,9,11,19,15]

fig,axs=plt.subplots(1,2,figsize=(8,4))
axs[0].bar(students,maths,color='b')
axs[0].set_title('Maths score')
axs[0].set_xlabel('students')
axs[0].set_ylabel('score')
axs[1].bar(students,science,color='r')
axs[1].set_title('science score')
axs[1].set_xlabel('students')
axs[1].set_ylabel('score')

plt.tight_layout()
plt.show()
```

```python
x=[10,20,30,40,50,60]
y=[1,2,7,4,5,6]
plt.subplot(2,2,1)
plt.plot(x,y,color='g')
plt.subplot(2,2,2)
plt.pie(y)
plt.subplot(2,2,3)
plt.bar(x,y,color='pink')
plt.subplot(2,2,4)
plt.pie(x)
plt.show()
```

                          TEXT PLOT

```python
x=[10,20,30,40,50,60]
y=[1,2,7,4,5,6]
plt.plot(x,y,color='g')
plt.text(20,3,'yes',color='black',bbox={"facecolor":"pink"})
```

annotation and legend


```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,3,4,7,11]

plt.plot(x,y)
plt.title('plot with general text')
plt.xlabel('X-Axis')
plt.label('Y-axts')
plt.annotate(r'Java',xy=(4,7),xytext=(2,10),arrowprops=dict(facecolor='yellow',shrink=0.10),fontsize=12,color='red')
```

CHANGING AXES

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 7, 11]

plt.plot(x, y)
plt.xlim(0, 10)
plt.ylim(0, 12)
plt.show()
```

 CHANGING THE VALUES IN AXIS

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 7, 11]

plt.plot(x, y)
plt.xticks([1,2,3,4,5],['a','b','c','d','e'])
plt.title("SAMPLE PLOT")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

SAVING A GRAPH

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 7, 11]

plt.plot(x, y)
plt.savefig("savesample.png")
plt.show()
```
