```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path ='E:\\DATASETS\\socialmedia.csv'
df = pd.read_csv(file_path)
print("Columns that we will analyze on are : ")
for i in df:
    print(i,",\t\t\t\t\t\t",end=" ")
```

```python
print("Starting 10 values : ")
df.head(10)
```

```python
plt.bar(df['Age'],df['Total Time Spent'],color=['plum'])
plt.xlabel("Age")
plt.ylabel("Total Time Spent(in minutes)")
plt.title("Age-wise total time spent on social media")
plt.show()
```

```python
fig,axs=plt.subplots(1,2,figsize=(10,4))
gender=df['Gender']
labels=['Male','Female','Others']
label=['TikTok','Instagram' ,'YouTube','Facebook']
axs[0].pie(gender.value_counts(),labels=labels,hatch=['**O','O.O', '.||.'])
axs[0].set_title("Gender-wise classification")
platform_counts = df['Platform'].value_counts()
axs[1].bar(platform_counts.index, platform_counts.values,color=['grey','black'])
axs[1].set_title('Platform Usage Distribution')
axs[1].set_xlabel('Platform')
axs[1].set_ylabel('Count')
plt.show()
```

```python
print("Maximum number of sessions : ", max(df['Number of Sessions']))
print("Minimum number of sessions : ",min(df['Number of Sessions']))
```

```python
plt.figure(figsize=(14, 8))
location_counts = df['Location'].value_counts()
plt.bar(location_counts.index, location_counts.values,color=['darkseagreen','darkolivegreen'])
plt.title('Location Distribution')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(5,5))
debt_counts = df['Debt'].value_counts()
plt.bar(debt_counts.index, debt_counts.values,color=['teal','lemonchiffon'])
plt.title('Debt Status')
plt.xlabel('Debt')
plt.ylabel('Count')
plt.show()
```

```python
os_counts = df['OS'].value_counts()
plt.plot(os_counts.index,os_counts.values)
plt.title('Operating System Distribution')
plt.xlabel('Operating System')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df['Self Control'],df['Addiction Level'])
plt.title('Self Control vs. Addiction Level')
plt.xlabel('Self Control')
plt.ylabel('Addiction Level')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
df.boxplot(column='Satisfaction', by='Platform')
plt.title('Satisfaction Levels by Platform')
plt.ylabel("Satisfaction")
plt.show()
```

```python
aggregated_data = df.groupby('Frequency')['Total Time Spent'].sum()
plt.figure(figsize=(10, 6))
plt.plot(aggregated_data.index, aggregated_data.values, marker='o')
plt.title('Total Time Spent by Frequency')
plt.xlabel('Frequency')
plt.ylabel('Total Time Spent')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(24, 6))
plt.barh(df['Video Category'],df['Engagement'],color=['midnightblue'])
plt.title('Average Engagement by Video Category')
plt.xlabel('Video Category')
plt.ylabel('Average Engagement')
plt.show()
```

```python
plt.figure(figsize=(24, 6))
label=['Comedy','Trends','Life Hacks','ASMR','Entertainment','Jokes/Memes','Gaming','Vlogs','P
plt.pie(df['Video Category'].value_counts(),labels=df['Video Category'],colors=['grey','black','darkseagreen','darkolivegreen','teal','lemonchiffon','plum','midnightblue','indigo'])
plt.title('Video Category wise distribution')
plt.show()
```
