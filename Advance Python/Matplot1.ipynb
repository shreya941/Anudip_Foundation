```python
import matplotlib.pyplot as plt

days=[1,2,3,4,5,6]
temperatures=[55,67,68,67,66,64]
plt.bar(days,temperatures,color=['red','green','pink','blue','yellow','orange'])
plt.xlabel("DAYS")
plt.ylabel("TEMPERATURE")
plt.title("Temperature of 5 days")
plt.show()
```

```python
import matplotlib.pyplot as plt

days=[1,2,3,4,5,6]
temperatures=[55,67,68,67,66,64]
plt.bar(days,temperatures,color='pink',width=0.3,alpha=0.5)
plt.xlabel("DAYS")
plt.ylabel("TEMPERATURE")
plt.title("Temperature of 5 days")
plt.show()
```
